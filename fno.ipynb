```python
import pandas as pd
import rarfile
import io
import warnings
```

```python
warnings.filterwarnings('ignore', category=FutureWarning)
with rarfile.RarFile('2023.rar', 'r') as rar_file:
    df = pd.DataFrame()
    for file_name in rar_file.namelist():
        if file_name.endswith('.csv'):
            file_content = io.StringIO(rar_file.read(file_name).decode())
            csv_df = pd.read_csv(file_content)
            csv_df = csv_df[(csv_df['SYMBOL'] == 'HDFC') & (csv_df['OPTION_TYP'] == 'CE')]
            df = df.append(csv_df, ignore_index=True)
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'], errors='coerce')
```

```python
df.info()
```

```python
#Close price by date as columns
close_by_date_view = pd.pivot_table(df, values='CLOSE', index='EXPIRY_DT', columns='TIMESTAMP')
close_by_date_view
```
